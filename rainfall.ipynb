### 컬럼 설명
- 'rainfall_train.fc_year' -> 기준 발표 연도
- 'rainfall_train.fc_month' -> 기준 발표 월
- 'rainfall_train.fc_day' -> 기준 발표 날짜
- 'rainfall_train.fc_hour' -> 기준 발표 시간
- 'rainfall_train.stn4contest' -> 지점
- 'rainfall_train.dh' -> 기준시각 - 예측 시간 
    - 기준발표 시각으로부터 240시간까지 3시간 간격으로 80개의 예측 자료의 시간을 의미
- 'rainfall_train.ef_year' -> 예측 시간 연도
- 'rainfall_train.ef_month' -> 예측 시간 월
- 'rainfall_train.ef_day' -> 예측 시간 날짜
- 'rainfall_train.ef_hour' -> 예측 시간 
- 'rainfall_train.v01' -> 0.1mm 이상 누적 확률
- 'rainfall_train.v02' -> 0.2mm 이상 누적 확률
- 'rainfall_train.v03' -> 0.5mm 이상 누적 확률
- 'rainfall_train.v04' -> 1.0mm 이상 누적 확률
- 'rainfall_train.v05' -> 2.0mm 이상 누적 확률
- 'rainfall_train.v06' -> 5.0mm 이상 누적 확률
- 'rainfall_train.v07' -> 10.0mm 이상 누적 확률
- 'rainfall_train.v08' -> 20.0mm 이상 누적 확률
- 'rainfall_train.v09' -> 30.0mm 이상 누적 확률
- 'rainfall_train.vv' -> 실제 강수량
    - 지점별 1시간 강수자료의 3시간 누적 값
    - 검증 데이터에서는 제외되어있음
- 'rainfall_train.class_interval' -> 강수 계급
    - 10단계의 강수 계급으로 검증파일(test)에서 -999로 표기된 부분은 결측으로 정답 파일에서 변경 없이 제출
